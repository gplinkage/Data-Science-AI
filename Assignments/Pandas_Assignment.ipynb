```python
#b-1. What is the likelihood that a male student feels that men and women can be just friends? (Report the proportion rounded to 2 decimal places.)

```

```python
male_count_for_yes = df[(df['Sex'] == 'male') & (df['Feasible'] == 'yes')].shape[0]
total_students = df.shape[0]
likelihood_male_yes = male_count_for_yes / total_students

likelihood_male_yes_rounded = round(likelihood_male_yes, 2)
likelihood_male_yes_rounded
```

```python
#b-2. What is the likelihood that a female student feels that men and women can be just friends? (Report the proportion rounded to 2 decimal places.)
```

```python
female_count_for_yes = df[(df['Sex']=='female') & (df['Feasible']=='yes')].shape[0]
total_students = df.shape[0]
likelihood_female_yes = female_count_for_yes/total_students

likelihood_female_yes_rounded = round(likelihood_female_yes,2)
likelihood_female_yes_rounded
```
