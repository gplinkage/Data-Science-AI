```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Data.csv')
df
```

```python
mean_salary = df['Salary'].mean().round(1)
mean_salary
```

```python
median_age = df['Age'].median().round(1)
median_age
```

```python
from sklearn.impute import SimpleImputer
impute_mean = SimpleImputer(strategy= 'mean')
impute_median = SimpleImputer(strategy= 'median')
```

```python
df['Salary'] = impute_mean.fit_transform(df[['Salary']]).round(0)
df
```

```python
df['Age'] = impute_median.fit_transform(df[['Age']])
df
```

```python
x= df.iloc[: , :3].values
y = df.iloc[:, -1].values
```

```python
x
```

```python
y
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [0])], remainder='passthrough')

x = np.array(ct.fit_transform(x))
x
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()

y = le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,  test_size = 0.2, random_state=0)
```

```python
x_train, x_test, y_train, y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()

x_train[:, 3:]= sc.fit_transform(x_train[:, 3:])
x_test[:, 3:]= sc.fit_transform(x_test[:, 3:])

x_train, x_test
```

```python
df=pd.read_csv('/content/Data3.csv')
df
```

```python
x= df.iloc[: , :5].values
y= df.iloc[:, -1].values
```

```python
x
```

```python
y
```

```python
from sklearn.impute import SimpleImputer
impute_mean = SimpleImputer(missing_values=np.nan, strategy = 'mean')
impute_median = SimpleImputer(missing_values=np.nan, strategy = 'median')
impute_mf = SimpleImputer(missing_values=np.nan, strategy = 'most_frequent')
impute_cost = SimpleImputer(missing_values=np.nan, strategy = 'constant', fill_value = 'India')
```

```python
x[:, 4:5] = impute_cost.fit_transform(x[:, 4:5])
x[:, 3:4] = impute_mf.fit_transform(x[:, 3:4])
x[:, 2:3] = impute_mean.fit_transform(x[:, 2:3]).round(0)
x[:, 1:2] = impute_median.fit_transform(x[:, 1:2])
```

```python
x
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct= ColumnTransformer(transformers= [('encoder', OneHotEncoder(), [0, 3, 4])], remainder= 'passthrough')
x= ct.fit_transform(x)
x
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y= le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state=42)
x_train, x_test, y_train, y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train[:, 7:] = sc.fit_transform(x_train[:, 7:])
x_test[:, 7:] = sc.fit_transform(x_test[:, 7:])
x_train, x_test
```
