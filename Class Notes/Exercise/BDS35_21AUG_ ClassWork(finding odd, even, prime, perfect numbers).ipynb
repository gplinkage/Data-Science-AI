```python
#1) Write a Python Function that accepts a hyphen-separated sequence of words as input and prints the words in a hyphen-separated sequence after sorting them alphabetically.
#Sample Items : green-red-yellow-black-white
#Expected Result : black-green-red-white-yellow
```

```python
# Define a function named "sort_the_string" that accepts a hyphen-separated string of words
def sort_the_string(words_str):
    # Split the input string into a list of words using the hyphen as a separator
    words_list = words_str.split("-")

    # Sort the list of words alphabetically
    sorted_list = sorted(words_list)

    # Join the sorted words back into a hyphen-separated string
    sorted_str = "-".join(sorted_list)

    # Print the sorted hyphen-separated string
    print(sorted_str)

# Example usage
input_string = "green-red-yellow-black-white"
sort_the_string(input_string)
```

```python
# 2)  Write a function that accepts a string and count consonant and vowels in that string

def counter1(string):
    # Initialize counters for consonants and vowels
    consonant_count = 0
    vowel_count = 0

    # Define sets of vowel characters
    vowels = "aeiouAEIOU"

    # Loop through each character in the input string
    for i in string:
        if i in vowels:
            vowel_count += 1  # Increment vowel count if the current character is in the vowel set
        elif i not in vowels and i!=" ":
            consonant_count += 1  # Increment consonant count if the current character is not in the vowel set

    # Print the results
    print("Consonant count:", consonant_count)  # Print the total number of consonants
    print("Vowel count:", vowel_count)  # Print the total number of vowels

# Example usage
input_text = "Hello World"
counter1(input_text)
```

```python
#4) Write a program to print twin primes less than 1000. If two consecutive odd numbers are both prime then they are known as twin primes
# logical breaking
# first part is we have to iterate till 1000
# 2nd part oly check for odd numbers
# 3rd both the numbers should be prime
```

```python
# Take input from the user for the maximum range
n = int(input("Enter the max range: "))

# Define a function to check if a number is prime
def check_prime_number(num):
    for i in range(2, num):
        if num % i == 0:
            return False  # If the number is divisible by any other number, it's not prime
    return True  # If no divisors found, the number is prime

# Define a function to print twin prime pairs
def print_prime_pairs(n):
    # Start from 3, the first odd prime number, and iterate by 2 to get the next odd numbers
    for k in range(3, n, 2):
        # Check if both k and k+2 are prime using the check_prime_number function
        if check_prime_number(k) and check_prime_number(k + 2):
            print("Pair Prime Numbers:", k, k + 2)

# Call the print_prime_pairs function with the user-input range
print_prime_pairs(n)
```

```python
#5) Write a program to find out the prime factorization of a number. Example: prime factorization  of 56 2 × 2 × 2 × 7
# divide the number until it becomes 1
#
# Take user input for the number to perform prime factorization on
n = int(input("Enter the number for prime factorization: "))

# Function to calculate prime factors of a number
def prime_factors(num):
    factors = []  # List to store prime factors
    divisor = 2   # Starting from the smallest prime divisor

    # Continue dividing 'num' by 'divisor' as long as 'num' is greater than 1
    while num > 1:
        while num % divisor == 0:
            num = num / divisor  # Update 'num' by dividing it
            factors.append(str(divisor))  # Add 'divisor' to the list of factors
        divisor = divisor + 1  # Move to the next divisor

    return factors  # Return the list of prime factors

# Function to display prime factorization in a readable format
def show_prime_factorization(n):
    factors = prime_factors(n)  # Get the prime factors using the previous function
    return " x ".join(factors)  # Join the factors with " x " in between for display

# Call the function to display the prime factorization
result = show_prime_factorization(n)
print("Prime Factorization:", result)
```

```python
# 6) Two different numbers are called amicable numbers if the sum of the proper divisors of each is equal to the other number. For example 220 and 284 are amicable numbers.
# get all divisors of both numbers
# check if sum is equal to original number

# Get two numbers from the user
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

# Function to calculate the sum of proper divisors of a number
def get_proper_divisors_sum(num):
    sum = 0  # Initialize the sum to 0
    for k in range(1, num):
        if num % k == 0:  # If 'k' divides 'num' evenly
            sum += k      # Add 'k' to the sum
    return sum          # Return the sum of proper divisors

# Function to check if two numbers are amicable (ambicale) numbers
def check_amicable_numbers(x, y):
    # Compare the sum of proper divisors of each number with the other number
    if get_proper_divisors_sum(x) == y and get_proper_divisors_sum(y) == x:
        print("These are amicable numbers:", x, y)
    else:
        print("Not amicable numbers")

# Call the function to check if the provided numbers are amicable

check_amicable_numbers(x, y)

```

```python
#1) Swap first and last element of the list
# a=[1,2,5,10,'1']

def swap_first_last(lst):
    if len(lst) >= 2:  # Check if the list has at least two elements
        lst[0], lst[-1] = lst[-1], lst[0]  # Swap the first and last elements

# Original list
a = [1, 2, 5, 10, '1']

# Call the function to swap first and last elements
swap_first_last(a)

# Print the list after swapping
print("List after swapping:", a)
```
