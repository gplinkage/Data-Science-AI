```python
#Program 1: Print following pattern
# Outer loop to control the number of rows
for i in range(10):
    # Inner loop to print leading spaces in each row
    for j in range(i+1):
        print('', end=' ')  # Print a space without a newline

    # Inner loop to print numbers in increasing order
    for k in range(i + 1, 11, 1):  # Start from i+1 and go up to 10
        if k == 10:
            print(0, end=' ')  # Print '0' instead of '10'
        else:
            print(k, end=' ')  # Print numbers from i+1 to 9

    print('\r')  # Move to the next line after printing the row
```

```python
# Program 2
# Outer loop: Loop through numbers 0 to 9 (total 10 iterations)
for i in range(10):
    # First nested loop: Print leading spaces
    for j in range(i + 1):
        print('', end=' ')

    # Second nested loop: Print numbers in descending order
    for k in range(10 - i, 0, -1):
        if k == 10:
            print(0, end=' ')  # Print '0' at the beginning of each row
        else:
            print(k, end=' ')   # Print the number

    print('\r')  # Move to the next line after each row is printed
```

```python
# Set the value of n (number of rows in the pattern)
n = 8

# Outer loop that controls the number of rows (k ranges from 1 to n)
for k in range(1, n + 1):
    # First inner loop to print numbers in increasing order
    for m in range(1, k):  # Loop to print values from 1 to k-1
        print(m, end=" ")  # Print the current value followed by a space

    # Second inner loop to print numbers in decreasing order
    for m in range(k, 0, -1):  # Loop to print values from k to 1 in reverse
        print(m, end=" ")  # Print the current value followed by a space

    print('\r')  # Move to the next line after printing the row
```

```python
 n = 10  # Define the value of n, which determines the number of rows in the pattern

# Outer loop: Iterate through each row (r represents the current row)
for r in range(1, n):  # The loop will run from 1 to n-1 (inclusive)
    digit = 1  # Initialize a variable to keep track of the value to print in each position

    # Inner loop: Iterate through each column (m represents the current column)
    for m in range(n, 0, -1):  # The loop will run from n to 1 (inclusive) in reverse
        if m > r:  # If the current column index is greater than the current row index
            print("", end=' ')  # Print a space (no content) with a space as the separator
        else:
            print(digit, end=' ')  # Print the current digit value with a space as the separator
            digit += 1  # Increment the digit value for the next position

    print("")  # Print a newline to move to the next row
```

```python
# Set the value of 'n' to 10
n = 10

# Outer loop: Iterate from 0 to (n - 1)
for k in range(0, n):  # 'k' represents the current row number
    # Inner loop: Iterate from 0 to 'k'
    for m in range(0, k + 1):  # 'm' represents the current column number in the row
        # Print the product of 'k' and 'm' with a space after
        print(k * m, end=' ')
    # Move to the next line after each row is printed
    print('\r')
```

```python
n = 10  # Set the value of 'n' to 10, which determines the number of rows in the pattern

# Outer loop: Controls the number of rows
for k in range(0, n):
    # First inner loop: Prints decreasing numbers before center
    for m in range(n - 1, k - 1):
        print(m, end=" ")  # Print the value of 'm' followed by a space

    # Second inner loop: Prints spaces for indentation
    for j in range(k):
        print('', end=" ")  # Print a space for indentation

    # Third inner loop: Prints increasing numbers after center
    for h in range(k + 1, n):
        print(h, end=" ")  # Print the value of 'h' followed by a space

    print('\r')  # Move to the next line after completing a row

```

```python
# Initialize variables for the beginning and end of the sequence
beg = 1
end = 2

# Initialize a variable to keep track of the current number
curr = end

# Outer loop to control the number of rows
for r in range(1, 8):
    # Inner loop to generate numbers in the current row
    for c in range(beg, end):
        curr -= 1  # Decrease the current number by 1
        print(curr, end=' ')  # Print the current number followed by a space
    print("")  # Move to the next line after printing the row

    # Update the beginning and end values for the next row
    beg = end
    end += r  # The end value increases with each row

    # Reset the current number for the next row
    curr = end
```
