```python
img1 = cv2.imread('/content/MicrosoftTeams-image (1).png')
img2 = cv2.imread('/content/MicrosoftTeams-image (2).png')
```

```python
cv2_imshow(img1)
cv2_imshow(img2)
```

```python
weightedsum = cv2.addWeighted(img1, 0.5, img2, 0.8, 0)
cv2_imshow(weightedsum)
```

```python
substracted = cv2.subtract(weightedsum, img2)
cv2_imshow(substracted)
```

```python
#Blurring
image_orig = cv2.imread('/content/fruit_example.jpg')

cv2_imshow(image_orig)
```

```python
#GaussianBlur # kernel values can not be even for height and witdth
Gaussian = cv2.GaussianBlur(image_orig, (5,5),0)
cv2_imshow(Gaussian)
```

```python
# median blurr
median = cv2.medianBlur(image_orig,3)
cv2_imshow(median)
```

```python
bilateral = cv2.bilateralFilter(image_orig, 9, 75, 75)
cv2_imshow(bilateral)
```
