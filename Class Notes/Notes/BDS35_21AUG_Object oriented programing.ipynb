```python
# Step 1: Define the Class
class ToyRobot:
    pass
# This creates a class named ToyRobot. Think of it as a blueprint.

# Step 2: Add the __init__ Method
class ToyRobot:
    def __init__(self):
        # This special method is like instructions for setting up the robot.
        # It runs automatically when a new object is created.
        self.color = "Red"   # The robot's color is set to "Red" initially.
        self.power = "Battery"  # The robot's power source is set to "Battery" initially.
# Create a toy robot object
my_robot = ToyRobot()

# Access the properties
print("Robot color:", my_robot.color)  # This prints the robot's color, which is "Red".
print("Robot power:", my_robot.power)  # This prints the robot's power source, which is "Battery".
```

```python
# Define the base class 'person'
class person:
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname

    def printname(self):
        print('class person:', self.firstname, self.lastname)
```

```python

# Define the child class 'childclass' that inherits from 'person'
class childclass(person):
    def __init__(self, fname, lname, address):  # Include fname and lname as parameters
        # Call the parent class's constructor using 'super()' to set the first and last name
        super().__init__(fname, lname)
        self.address = address

    def address1(self):
        print(self.address)
```

```python
# Create an instance of the base class 'person'
obj1 = person('John', 'Smith')
```

```python
# Create an instance of the child class 'childclass'
obj2 = childclass('John', 'Smith', '22, ABC Street, NY 5088')  # Provide fname, lname, and address
```

```python
# Call the 'printname' method of the base class
obj1.printname()
```

```python
# Call the 'printname' method of the child class (inherited from base class)
obj2.printname()
```

```python
# Call the 'address1' method of the child class
obj2.address1()
```

```python
class Car:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Drive!")

class Boat:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Sail!")

class Plane:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Fly!")

car1 = Car("Ford", "Mustang")       #Create a Car class
boat1 = Boat("Ibiza", "Touring 20") #Create a Boat class
plane1 = Plane("Boeing", "747")     #Create a Plane class

for x in (car1, boat1, plane1):
  x.move()
```

```python
class vehicle:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print('move!!')

class Car(vehicle):
  def move(self):
    print("Drive!")

class Boat(vehicle):
  def move(self):
    print("Sail!")

class Plane(vehicle):
  def move(self):
    print("Fly!")

car1 = Car("Ford", "Mustang")       #Create a Car class
boat1 = Boat("Ibiza", "Touring 20") #Create a Boat class
plane1 = Plane("Boeing", "747")     #Create a Plane class

for i in (car1,boat1,plane1):
  print(i.brand)
  print(i.model)
  i.move()
  print("========")
```

```python
class Animal:
    def __init__(self, name, species,friendliness):
        self.name = name
        self.species = species
        self.friendliness = friendliness

    def eats(self):
        print("Animal eats")

class Dog(Animal):
    def eats(self):
        print("Dog eats bones")

class Cat(Animal):
    def eats(self):
        print("Cat eats tuna")

dog1 = Dog("Teddy", "Golden Retriever", "Very Friendly")
cat1 = Cat("Blue", "Gray", "Moderately Friedly")

for i in (dog1, cat1):
    print(i.name)
    print(i.species)
    print(i.friendliness)
    i.eats()
    print("================")
```

```python
class vehicle():
  def __init__ (self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Vehicle moves")

class car(vehicle):
  def move (self):
     print("Car drives")

class boat(vehicle):
  def move (self):
     print("Boat Sails")

class plane(vehicle):
  def move (self):
     print("Plane Flys")

car1= car("Toyota", "Camry")
boat1= boat("Suzuki", "Swift")
plane1 = plane ("Boeing", "380")

for i in (car1, boat1, plane1):
    print(i.brand)
    print(i.model)
    i.move()
    print("===========")
```

```python
#Class practice code for Inheritance, Polymorphism & inheritance:

class Bootcamp:
  def __init__(self, name):
    self.name = name

def role(self):
  pass

class Instructor(Bootcamp):
  def role(self):
      return "Teaches the curriculum and grades the students"

class Organizer(Bootcamp):
  def role(self):
    return "Organizes the bootcamp and resolves issues."

class Student(Bootcamp):
   def role(self):
      return "Attends the bootcamp and completes assignments"

instructor1 = Instructor("Anurag")
organizer1 = Organizer("Nirusa")
student1 = Student("Pradeep")

roles = [instructor1, organizer1, student1]
for r in roles:
  print(r.name, end=' - ')
  print(r.role())
  r.role()
```
