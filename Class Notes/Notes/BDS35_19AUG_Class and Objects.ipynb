```python
# Define a class named Human
class Human:
    def __init__(self, n, a):
        # Initialize instance variables name and occupation
        self.name = n
        self.occupation = a

    # Method to perform work based on occupation
    def do_work(self):
        if self.occupation == "Actor":
            print(self.name, "Plays roles in movies")  # Print actor's work description

        elif self.occupation == "Tennis Player":
            print(self.name, "Plays tennis")  # Print tennis player's work description

    # Method to make the human speak
    def speaks(self):
        print(self.name, "Says, 'How are you?'")  # Print a generic speaking phrase

# Create an instance of the Human class for Tom Cruise
tom = Human("Tom Cruise", "Actor")
tom.do_work()  # Call the do_work method for Tom
tom.speaks()   # Call the speaks method for Tom

# Create an instance of the Human class for Maria Sharapova
maria = Human("Maria Sharapova", "Tennis Player")
maria.do_work()  # Call the do_work method for Maria
maria.speaks()   # Call the speaks method for Maria
```
