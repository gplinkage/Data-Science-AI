# Hierarchical Clustering

## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
df = pd.read_csv('/content/Mall_Customers.csv')
X= df.iloc[:, [3,4]].values
X
```

## Using the dendrogram to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dndogram = sch.dendrogram(sch.linkage(X, method = 'ward'))

plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.tight_layout()
plt.show()
```

## Training the Hierarchical Clustering model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc= AgglomerativeClustering(n_clusters= 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

## Visualising the clusters

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot for Cluster 1
sns.scatterplot(x=X[y_hc == 0, 0], y=X[y_hc == 0, 1], s=100, color='red', label='Cluster 1')

# Scatter plot for Cluster 2
sns.scatterplot(x=X[y_hc == 1, 0], y=X[y_hc == 1, 1], s=100, color='blue', label='Cluster 2')

# Scatter plot for Cluster 3
sns.scatterplot(x=X[y_hc == 2, 0], y=X[y_hc == 2, 1], s=100, color='green', label='Cluster 3')

# Scatter plot for Cluster 4
sns.scatterplot(x=X[y_hc == 3, 0], y=X[y_hc == 3, 1], s=100, color='cyan', label='Cluster 4')

# Scatter plot for Cluster 5
sns.scatterplot(x=X[y_hc == 4, 0], y=X[y_hc == 4, 1], s=100, color='magenta', label='Cluster 5')

# Title and labels for the plot
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')

# Display legend to show cluster labels
plt.legend()

# Show the plot
plt.show()

```
